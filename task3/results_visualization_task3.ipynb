__Subtasks 4. Оцінювання текстової семантики на португальській__

Оцінка семантики текстових коментарів на португальській здійснюється за допомогою коду, який призначений для визначення значення тексту та надання йому оцінки від 1 до 5.

Для вирішення даної задачі була обрана модель __RoBERTa__ через її найкращі метрики продуктивності.   
Модель була навчена на понад 43 000 прикладах. В результаті роботи досягається точність на рівні 71%.  

Також варто відзначити, що базова логістична регресія показала точність на рівні 54%.   
Базовий BERT і BERT для португальської також показали гірші результати порівняно з RoBERTa, що підкреслює ефективність вибору RoBERTa для цієї задачі.

```python
import torch
from transformers import RobertaForSequenceClassification, RobertaTokenizer

class SentimentClassifier:
    def __init__(self, model_path):
        self.tokenizer = RobertaTokenizer.from_pretrained('roberta-base')
        self.model = RobertaForSequenceClassification.from_pretrained(model_path)

    def predict_sentiment(self, text):
        inputs = self.tokenizer(text, padding="max_length", truncation=True, max_length=128, return_tensors='pt')
        outputs = self.model(**inputs)
        predicted_class = torch.argmax(outputs.logits, dim=1).item()
        predicted_score = predicted_class + 1
        return predicted_score

model_path = "new_data_task3/model_training_results/checkpoint-6500"
sentiment_classifier = SentimentClassifier(model_path)
```

```python
text = " Boa noite! Gostaria de saber se tem como eu cancelar esse pedido, não posso mais esperar. Obrigado"
predicted_score = sentiment_classifier.predict_sentiment(text)
print("Передбачений рейтинг:", predicted_score)
```
