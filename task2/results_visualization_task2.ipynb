__Subtasks 2. Лідери і аутсайдери за продажами у кожній області.__

Висновки щодо бізнес-цінності аналітичного звіту лідерів і аутсайдерів за продажами у кожній області:  
__Виявлення монопольного впливу:__ В областях AC, AM, MA, PA, PI існує лише один продавець, що дає йому монопольний статус.  
__Аналіз лідерів та аутсайдерів:__ Графік вказує на тенденцію, за якою зі зменшенням загального обсягу продажів знижується обсяг продажів лідерів (за винятком штатів BA, PE, PB), а обсяг продажів аутсайдерів зростає (окрім штатів DF, MT). Це може вказувати на зміну уподобань споживачів або покращення конкурентоспроможності менших компаній. Варто детальніше дослідити штати DF, MT, тому що відхилення свідчить про те, що у цих штатах не з'являються нові продавці.  
__Стратегічне планування:__ Дані підкреслюють потребу у розробці стратегій залучення та підтримки нових продавців у штатах, де спостерігається монопольне становище або домінування аутсайдерів. Зокрема, варто встановити відсоток домінації лідерів не вище 50%. Це може включати стимулювання конкуренції та вдосконалення умов для входу на ринок.  
__Оптимізація ресурсів:__ Розуміння динаміки між лідерами та аутсайдерами може допомогти компанії краще розподілити ресурси, вибираючи, де інтенсифікувати маркетингові та продажні зусилля для максимізації рентабельності. 
__Ринкова диференціація:__ Варто провести детальніший аналіз для ідентифікації унікальних стратегічних відмінностей, що може надати нові можливості.  
__Залежність від одного регіону:__ Об'єм продажів у регіоні SP (64% від загального обсягу) свідчить про велику залежність від цього конкретного регіону. Варто детальніше дослідити ринок, тому що це може мати як негативний вплив, так і позитивний. Як наслідок, потрібно проводити диверсифікацію.

Висновки щодо бізнес-цінності аналітичного звіту товарів, які продаються найкраще в кожній категорії:  
__Ефективність асортименту товарів:__ Виявлено, що в деяких категоріях невелика кількість товарів або навіть один продукт може забезпечити значний обсяг продажів. Це свідчить про важливість аналізу популярності конкретних товарів у кожній категорії та оптимізації асортименту. "cds_dvds_musicals" категорія в якій всього лише 1 товар.

__Стратегічне розподілення ресурсів:__ Фокусування уваги на категоріях з великим обсягом продажів може бути ключовим для максимізації прибутковості.
