```python
import pandas as pd

products_df = pd.read_csv("../data/products.csv")
order_items_df = pd.read_csv('../data/order_items.csv')
```

```python
order_items_df = order_items_df.drop_duplicates(subset=['order_id', 'product_id'], keep='first')

merged_df = pd.merge(order_items_df, products_df, on='product_id', how='left')

columns_to_drop = ['shipping_limit_date', 'product_name_lenght', 'order_item_id', 'seller_id', 'product_category_name', 'product_description_lenght', 'product_photos_qty']
merged_df.drop(columns=columns_to_drop, inplace=True)
print(len(merged_df['order_id']))

merged_df.head()
```

```python
correlation_matrix = merged_df[['product_weight_g', 'product_length_cm', 'product_height_cm', 'product_width_cm', 'freight_value', 'price']].corr()
correlation_df = pd.DataFrame(correlation_matrix)

correlation_df.drop('freight_value', axis=0, inplace=True)
correlation_df.index.name = 'parameters'

file_path = 'new_data_task2/dependence_of_product_parameters_on_turnover_and_price.csv'
correlation_df.to_csv(file_path, index=True, header=True)

correlation_df.head(20)
```
