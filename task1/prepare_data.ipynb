```python
import pandas as pd

translations_df = pd.read_csv("../data/product_category_name_translation.csv")
products_df = pd.read_csv("../data/products.csv")
order_items_df = pd.read_csv('../data/order_items.csv')
order_df = pd.read_csv('../data/orders.csv')
```

```python
merged_df = pd.merge(products_df, translations_df, how='left', on='product_category_name')
merged_df.head()
```

```python
order_items_df.drop_duplicates(subset=['order_id', 'product_id'], keep='first', inplace=True)
print(len(order_items_df['order_id']))

merged_df = pd.merge(order_items_df, merged_df, on='product_id', how='left')

print(len(merged_df['order_id']))
merged_df.head()
```

```python
merged_order_df1 = pd.merge(order_df, merged_df, on='order_id', how='left')
print(len(merged_order_df1['order_id']))

columns_to_drop = ['order_status','shipping_limit_date', 'order_item_id', 'seller_id', 'order_id', 'customer_id', 'product_category_name', 'product_id', 'order_approved_at', 'order_delivered_carrier_date', 'order_delivered_customer_date', 'order_estimated_delivery_date']
merged_order_df1.drop(columns=columns_to_drop, inplace=True)

print(len(merged_order_df1['order_purchase_timestamp']))

merged_order_df1 = merged_order_df1.dropna(subset=['price', 'product_category_name_english', 'order_purchase_timestamp'], how='any')

print(len(merged_order_df1['order_purchase_timestamp']))

file_path = 'new_data_task1/merged_order_df1.csv'
merged_order_df1.to_csv(file_path, index=False)

merged_order_df1.head()
```
